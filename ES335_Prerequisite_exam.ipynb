```python
# question 3 
# implementing a code to find the second highest element by succesive method 
list = [1,2,3,4,5,10,4,8,9, 11, 12, 15, 10, 25 ,21, 30]  
max = list[0]
for i in range(0,len(list)):
  if list[i]>max:
    max=list[i]
  else : 
    continue 
max_second = list[0]
for i in range(0,len(list)):
  if list[i]<max and list[i]> max_second:
    max_second=list[i]
  else : 
    continue 
print("Maximum : ", max)
print("Second Maximum : ",max_second)

# Similary median would be kth highest element where k = (n)/2 if n is odd and ((n/2)-1) if n is even 
def find_kth_max(list,max_current):
 max = list[0]
 for j in range(0,len(list)):
  if list[j]>max and list[j] < max_current: 
    max_current = list[j]
  else: 
    continue 
  return max_current 

def median(list):
  n = len(list)
  if n%2 == 0:
    k = ((n/2)-1)
  else :
    k = (n+1)//2
  max_current = list[0]
  for i in range(0,k):
    max_current = find_kth_max(list,max_current)
  return max_current 

list = [1,2,3,4,5,10,4,8,9, 11, 12, 15, 10, 25 ,21, 30]  
median = median(list)
print(median)
```

```python
# Question 4 
# Analytically solving without libraries 
def partial_derivatives(x,y):
  df_dx = 2*x*y + y*y*y*sp.cos(x)
  df_dy = x**2 + 3*y**2*sp.sin(x)
  return df_dx, df_dy
  
gradient = partial_derivatives(x,y)
print("Gradient : ",gradient)
```

```python
#question 6 
# Gradient of a function is the partial derivative of the function with respect to both the variables x and y
# Hence partial derivative of the function is : 
# df/dx|y = 2xy + y^3*cosx
# df/dy|x = x^2 + 3y^2*sinx
# Using sympy to find [gradient  = (df/dx|y, df/dy|x) 
import sympy as sp 
import numpy as np
x, y = sp.symbols('x y', real=True)
f = x*x*y + y*y*y*sp.sin(x)
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)
gradient = sp.Matrix([df_dx, df_dy])
print("Gradient : ", gradient)
```

```python
#Question 9 
# using matplotlib 
# plotting y = x 
import matplotlib.pyplot as plt 
import numpy as np

x = np.linspace(0,1,1000)
def f1(x):
  return x;
def f2(x):
  return x**2;
def f3(x):
  return x**3/100;
def f4(x):
  return np.sin(x);
def f5(x):
    return np.where(x != 0, np.sin(x) / x, np.nan)
def f6(x):
  return np.where(x>0,np.log(x),np.nan);
def f7(x):
  return np.exp(x);
y1 = f1(x)
y2 = f2(x)
y3 = f3(x)
y4 = f4(x)
y5 = f5(x)
y6 = f6(x)
y7 = f7(x)
plt.plot(x, y1, label='f1(x) = x')
plt.plot(x, y2, label='f2(x) = x^2')
plt.plot(x, y3, label='f3(x) = x^3 / 100')
plt.plot(x, y4, label='f4(x) = sin(x)')
plt.plot(x, y5, label='f5(x) = sin(x) / (x)')
plt.plot(x, y6, label='f6(x) = log(x)')
plt.plot(x, y7, label='f7(x) = exp(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
